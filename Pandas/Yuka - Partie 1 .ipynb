# Mission 3

Crée une fonction qui prend en paramètre un dataframe, et affiche le taux de remplissage de chacune de ces colonnes sous forme de pourcentage. C'est à dire :
- colonne 1 : 73%
- colonne 2 : 99%

```
def remplissage(df):
  compt = 0
  for col in df.columns:
    compt += 1
    print(f"colonne {compt} : {round(100 - ((df[col].isna().sum()/df.shape[0])*100), 2)}%")
```

```
remplissage(df)
```

# Mission 4

Quel est le nombre de colonne appartenant à chaque quartile des taux de remplissage ? Autrement dit, combien de colonne sont remplies à 0%, entre 0 et 25%, entre 25 et 50%...

```
# création d'une fonction qui me renvoie une liste des taux de remplissage des colonnes
def remplissage_list(df):  
  rempl= []
  for col in df.columns:
    rempl.append(round(100 - ((df[col].isna().sum()/df.shape[0])*100), 2))
  return rempl
```

```
# fonction quartile
def quartile_0(x):
  return x == 0

def quartile_1(x):
  return 0 < x < 25

def quartile_2(x):
  return 25 <= x < 50

def quartile_3(x):
  return 50 <= x < 75

def quartile_4(x):
  return 75 <= x <= 100
```

```
# liste
liste_remplissage= remplissage_list(df)

# nombre de colonne dans chaque quartile
print("Nombre de colonnes à 0% de remplissage :", sum(quartile_0(x) for x in liste_remplissage))
print("Nombre de colonnes entre 0% et 25% de remplissage :", sum(quartile_1(x) for x in liste_remplissage))
print("Nombre de colonnes entre 25% et 50% de remplissage :", sum(quartile_2(x) for x in liste_remplissage))
print("Nombre de colonnes entre 50% et 75% de remplissage :", sum(quartile_3(x) for x in liste_remplissage))
print("Nombre de colonnes entre 75% et 100% de remplissage :", sum(quartile_4(x) for x in liste_remplissage))
```
