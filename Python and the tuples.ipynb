# Mission 4

Create 2 tuples that contain only integers.
Create a script that compares them to each other and indicates which one of the tuples returns the highest total value after adding all their values together. 

```
tuple1 = (12, 126, 5, 89, 67, 51, 36)
tuple2= (23, 58, 96, 1, 4, 22)
compteur1 = 0
compteur2 = 0

for i in tuple1 :
  compteur1 += i

for i in tuple2:
  compteur2 += i

if compteur1 == compteur2:
  print('Les totaux des deux tuples sont égaux')
elif compteur1 > compteur2:
  print('Le total du tuple 1 est plus élevé')
else:
  print('Le total du tuple 2 est plus élevé')
```

# Mission 5

Here's a tuple:

my_tuple = ("data analyst", "data scientist", "data engineer", "data architect")

Create a script that allows you to change one of the values of "my_tuple" by indicating the position of the value you want to change, as well as the new value.

```
my_tuple = ("data analyst", "data scientist", "data engineer", "data architect")

value_to_change = int(input('Quelle est la position de la valeur à changer?'))

if value_to_change not in range(len(my_tuple)+1) or value_to_change == 0:
  print("merci d'indiquer un chiffre entre 1 et {}".format(len(my_tuple)))
  value_to_change = int(input('Quelle est la position de la valeur à changer?'))

new_value= input('Par quelle valeur tu veux la remplacer?')

# on passe le tuple en liste
y = list(my_tuple)

# modification à l'index - 1 pour coller à la position demandée
y[value_to_change - 1] = new_value

# on remet en type tuple
my_tuple = tuple(y)

print(my_tuple)
```
