```
# Mission 9: Create a function that takes a day as a parameter and returns the next day (i.e. for Sunday it would return Monday). 
# You can use lists or dictionaries to solve this.

days = ["monday", 'tuesday', 'wednesday','thursday', 'friday', 'saturday', 'sunday']

def next_day(day):
  day = day.lower()
  index_to_find = days.index(day)
  if index_to_find != 6:
    return days[index_to_find + 1]
  else:
    return days[0]

day_choice = input("Day of the week: ")
print(next_day(day_choice))
```

```
# Mission 10: Create a function that fills an empty list with values (words) from the following string : 
# "After twelve soft showers are the arch-duchess' socks dry, arch-dry?"

def fill_the_list(string):
  list_to_fill = string.split()
  return list_to_fill

sentence = "After twelve soft showers are the arch-duchess' socks dry, arch-dry?"
print(fill_the_list(sentence))
```

```
# Mission 11: Create a function that takes a str as a parameter, and returns the same string, where every other letter
# is written in uppercase, unless it's a consonant.

def other_letter_upper(sentence):
    sentence = sentence.lower()                               # si jamais des consonnes sont écrites en majuscule dans l'input()
    new_sentence = ''
    index = 0
    for letter in sentence:
        if index % 2 != 0 and letter in 'aeiouyAEIUOY':
            new_sentence += letter.upper()
            index += 1
        else:
            new_sentence += letter
            index += 1
    return new_sentence

phrase = input('Choose a word or a sentence: ')
print(other_letter_upper(phrase))
```

```
# Mission 12: Create a function that fills an empty list with string values of each letter of the following string: 
# "Python fan", and then displays the filled list

def fill_the_list2(string):
  new_string= string.replace(' ', '')
  list_to_fill = []
  for char in new_string:
    list_to_fill.append(char)
    
  return list_to_fill

sentence = "Python fan"
print(fill_the_list2(sentence))
```

```
# Mission 13: Create a function that take 2 inputs from the user, their date of birth and first name.
# Convert the date of birth into an int and subtract it from the current year (you can do it using a certain module :-P).
# Then display the message: Hello [first name], today you are [age] years old.
from datetime import *

def give_me_your_age(name, birthDate):
  today =  date.today()
  birthDate = datetime.strptime(birthDate, "%Y/%m/%d")
  age = today.year - birthDate.year - ((today.month, today.day) < (birthDate.month, birthDate.day))       #pour les personnes nées une année bisextile
  return 'Hello {}, today you are {} years old.'.format(name, age)

firstname = input("What's your name? ")
birthDate = input("What is your date of birth? (format YYYY/MM/DD) ")
print(give_me_your_age(firstname, birthDate))
```
