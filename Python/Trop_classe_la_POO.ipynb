- attributs :nom (str) prenom (str) age (int) sexe (str) formationPrecedente (str) motivation (int valant 100 à l'instanciation) progression (int valant 0 à l'instanciation)


- méthodes : 
* suivreFormation() : chaque fois qu'elle est utilisée, cette méthode diminue la motivation d'une valeur entière aléatoire (random) entre 5 et 25 et augmente la progression d'une valeur entière aléatoire entre 20 et 30 ; 
* bosserPlus() : chaque fois qu'elle est utilisée, cette méthode augmente la motivation et la progression d'une valeur entière aléatoire entre 10 et 30 ; 
* echouer() : chaque fois qu'elle est utilisée baisse la motivation d'une valeur entre 20 et 40 et augmente la progression d'une valeur entre 20 et 40 ;
* reussir() : chaque fois qu'elle est utilisée, augmente la motivation d'une valeur entre 20 et 40 et augmente la progression d'une valeur random entre 10 et 20

```python
import random

class DataAnalyst:

  def __init__(self, nom, prenom, age, sexe, formationPrecedente, motivation = 100, progression = 0):
    self.nom = nom
    self.prenom = prenom
    self.age = age
    self.sexe = sexe
    self.formationPrecedente = formationPrecedente
    self.motivation = motivation
    self.progression = progression
    print(f"Bonjour {prenom}")
  
  def get_motivation(self):
    return self.motivation

  def get_progression(self):
    return self.progression

  def suivreFormation(self):
    less_motiv = random.randint(5, 25)
    plus_progress = random.randint(20, 30)
    self.motivation -= less_motiv 
    self.progression += plus_progress
    print(f"La méthode suivreFormation() a été utilisée")
    print(f"Aie.. tu as perdu {less_motiv} points de motivation ! Il t'en reste: {DataAnalyst.get_motivation(self)}." )
    print(f"Mais c'est cool, tu as gagné {plus_progress} points de progression ! Tu en as désormais {DataAnalyst.get_progression(self)}.")

  def bosserPlus(self):
    plus_motiv = random.randint(10, 30)
    plus_progress = random.randint(10, 30)
    self.motivation += plus_motiv 
    self.progression += plus_progress
    print(f"La méthode bosserPlus() a été utilisée")
    print(f"Cool, tu as gagné {plus_motiv} points de motivation et {plus_progress} points de progressions!")
    print(f"Tu as désormais {DataAnalyst.get_motivation(self)} de motivation et {DataAnalyst.get_progression(self)} de progression.")
  
  def echouer(self):
    less_motiv = random.randint(20, 40)
    plus_progress = random.randint(20, 40)
    self.motivation -= less_motiv 
    self.progression += plus_progress
    print(f"La méthode echouer() a été utilisée")
    print(f"Aie.. tu as perdu {less_motiv} points de motivation ! Il t'en reste: {DataAnalyst.get_motivation(self)}." )
    print(f"Mais c'est cool, tu as gagné {plus_progress} points de progression ! Tu en as désormais {DataAnalyst.get_progression(self)}.")

  def reussir(self):
    plus_motiv = random.randint(20, 40)
    plus_progress = random.randint(10, 20)
    self.motivation += plus_motiv 
    self.progression += plus_progress
    print(f"La méthode bosserPlus() a été utilisée")
    print(f"Cool, tu as gagné {plus_motiv} points de motivation et {plus_progress} points de progressions!")
    print(f"Tu as désormais {DataAnalyst.get_motivation(self)} de motivation et {DataAnalyst.get_progression(self)} de progression.")

```

```python
data_analyst1 = DataAnalyst('Houssin','Alex', 31, "F", "M2 Finance")
```

```python
tour = 0
while data_analyst1.motivation > 0 or data_analyst1.progression < 100:
  tour += 1
  print(f'Tour: {tour}')
  random.choice((data_analyst1.suivreFormation, data_analyst1.bosserPlus, data_analyst1.echouer, data_analyst1.reussir))()
  print()
  if data_analyst1.motivation <= 0:
      print("BRAVO TU AS GAGNE !") 
      break     
  elif data_analyst1.progression >= 100:
      print("BRAVO TU AS APPRIS !")
      break
```
