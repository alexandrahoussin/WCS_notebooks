```
# Square of a number: Create a function that returns the square of a number

def square(nb):
  return nb ** 2

number = int(input('Choose a number: '))
print("The square is: {}.".format(square(number)))
```

```
# Cube of a number: Create a function that returns the cube of a number.
def cube(nb):
  return nb ** 3

number = int(input('Choose a number: '))
print("The cube is: {}.".format(cube(number)))
```

```
# Absolute value: Create a function that returns the absolute value of a number.
def abs_value(nb):
  if nb > 0:
    return nb
  else:
    return -nb

number = int(input('Choose a number: '))
print("The absolute value is: {}.".format(abs_value(number)))
```

```
# Factorial of a number: Create a function that returns the factorial of a number.

def factorial_number(nb):
  factorial = 1
  if nb < 0 :
    return "The factorial of a negative number doesn't exist"
  elif nb == 0:
    return factorial
  else:
    for i in range(1, nb+1):
      factorial *= i
    return factorial

number = int(input('Choose a number: '))
print("The factorial is: {}.".format(factorial_number(number)))
```

```
# Mode of a list of numbers: Create a function that returns the mode of a list of numbers.

def mode(list_of_nb):
  list_of_nb.sort()                                           # I sort the given list
  count_list = []                                             # create a new list where i'll put the occurences for each numbers
  i = 0
  while i < len(list_of_nb):
    count_list.append(list_of_nb.count(list_of_nb[i]))
    i += 1
  dict_of_occurence = dict(zip(list_of_nb, count_list))       # create a dict where keys = the number and values = how much we've seen this number in the list
  for key, value in dict_of_occurence.items():
    if value == max(count_list):                              # i'm searching the key where its values == the max of the count_list (the max nb of occurences)
      return key

my_list = [1, 5, 2, 3, 3, 1, 5, 8, 9, 3, 4, 6, 8]
print("The mode is: {}.".format(mode(my_list)))
```

```
# Average of a list of numbers: Create a function that returns the average of a list of numbers.

def average(list_of_nb):
  return sum(list_of_nb) / len(list_of_nb)

my_list = [1, 5, 2, 3, 3, 1, 5, 8, 9, 3, 4, 6, 8]
print("The average is: {}.".format(average(my_list)))

```

```
# Minimum of a list of numbers: Create a function that returns the minimum from within a list of numbers.

def minimum_nb(list_of_nb):
  min_number = list_of_nb[0] 
  for i in list_of_nb:
    if i < min_number:
      min_number= i
  return min_number

the_list = [52, 23, 14, 89, 6, 13, 475, 683, 9]
print("The minimum is: {}.".format(minimum_nb(the_list)))

```

```
# Maximum of a list of numbers: Create a function that returns the maximum from within a list of numbers.

def maximum_nb(list_of_nb):
  max_number = list_of_nb[0] 
  for i in list_of_nb:
    if i > max_number:
      max_number= i
  return max_number

the_list = [52, 23, 14, 89, 6, 13, 475, 683, 9]
print("The maximum is: {}.".format(maximum_nb(the_list)))
```
