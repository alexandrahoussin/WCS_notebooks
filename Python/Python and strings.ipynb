# Mission 2

Create a script that takes a string as one parameter and a character as another parameter. Your script should return the position (starting with 1 and not 0) of all occurrences of the character within the string.

For example, for the word "hello" and for the character "o", the script will return :
- "position : 5"

```
first_string = 'hello'
character = 'o'

# utilisation de la fonction enumerate pour obtenir l'index de chaque caractère de ma first_string
for i, char in enumerate(first_string):
  if char == character:
    print('position: {}'.format(i + 1))
```

# Mission 3

Create a script that only displays a particular part of a string it receives. In other words, make it so that you are able to decide the exact character range (e.g. 2nd to 5th character) to be displayed.

```
my_string = input('Entrez un mot ou une phrase ')
rang_one = int(input("Début de l'intervalle: "))
rang_end = int(input("fin de l'intervalle: "))

print('Voici le résultat: {}'.format(my_string[rang_one : rang_end]))
```

# Mission 4

Find 3 ways to return the following string without including the spaces at the beginning and the end: " Data Analyst ".

```
str1 = " Data Analyst "

# 1.
result1 = str1.strip()
print(result1)

# 2.
result2 = ' '.join(str1.split())
print(result2)

# 3.
result3 = str1.lstrip().rstrip()
print(result3)

```

# Mission 5

Create a script that converts all vowels to uppercase and all consonants to lowercase.

```
my_word = input('Entrez un mot: ')
answer = ""

for letter in my_word:
  if letter in "aeiouyAEIOUY":
    answer += letter.upper()
  else:
    answer += letter.lower()

print(answer)
```

# Mission 6

Create a script that replaces one character with another and redisplays the modified string.

```
my_word = input('Entrez un mot: ')
char_to_change = input('Quelle lettre souhaitez vous modifier? ')
new_char = input('Par quelle lettre souhaitez-vous la changer? ')

print('Votre nouveau mot: {}'.format(my_word.replace(char_to_change, new_char)))

```
