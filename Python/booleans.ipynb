```python
# saisir les bornes min et max
limit1, limit2 = input('Saisissez deux valeurs entières (séparées par des espaces): ').split()

# saisir l'entier à borner
int_to_limit = int(input('Saisissez un entier à borner: '))

# définir le minimum et le maximum
if limit1 < limit2:
  max = int(limit2)
  min = int(limit1)
else:
  max = int(limit1)
  min = int(limit2)


# comparaisons entre les limites et l'entier à borner
if min < int_to_limit < max:
  if (int_to_limit - min) < (max - int_to_limit):
    result = min
  else:
    result = max
elif int_to_limit >= max:
  result = max
else:
  result = min

print(result)
```
