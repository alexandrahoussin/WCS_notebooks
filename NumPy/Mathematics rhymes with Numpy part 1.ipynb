# Mission 1

Find a way to calculate the following statistical indicators for any given one-dimensional array using NumPy functions, and use them on a array filled with 20 random floats.
- mode
- median
- average
- the range (the difference between max and min)
- standard deviation
- variance
- the different quantiles

Then, create another array with 20 random floats and calculate the Pearson coefficient between the two arrays.

```
import numpy as np
```

```
arr1 = np.random.rand(20)
arr1
```

```
# MODE

# utilisation de np.unique() qui retourne les valeurs unique de l'array, ajout de return_counts = True pour compter le nombre de fois qu'elles apparaissent 
value, count = np.unique(arr1, return_counts= True)

# on recherche la valeur max dans le compteur avec argmax qui renvoie l'index de cette valeur
index = np.argmax(count)

# je retourne la valeur à cet index
value[index]
```

```
# vérification avec scipy
from scipy import stats

mode = stats.mode(arr1)
print(mode[0])
```

```
# MEDIAN
np.median(arr1)
```

```
# AVERAGE
np.average(arr1)
```

```
# RANGE
range = np.amax(arr1) - np.amin(arr1)
range
```

```
# STANDARD DEVIATION
np.std(arr1)
```

```
# VARIANCE
np.var(arr1)
```

```
# DIFFERENT QUANTILES
print(np.quantile(arr1, 0.25))
print(np.quantile(arr1, 0.5))
print(np.quantile(arr1, 0.75))
```

```
arr2 = np.random.rand(20)
print(arr1)
print(arr2)

# COEFF PEARSON
coeff_array = np.corrcoef(arr1, arr2)
print(coeff_array)
```

## Mission 2

Create a function using NumPy that indicates the values (integers only), from -100 to 100, where the remainder from performing Euclidean division by both 3 and 7 is equal.

```

def remainder_euclidean(array):
  # création de 2 array (un des reste de division par 3 et un des restes de division par 7)
  remainder_by_3 = np.remainder(array, 3)
  remainder_by_7 = np.remainder(array, 7)

  #création liste vide où je vais stocker les index des valeurs égales dans les arrays
  list_vide = []
  for i in range(len(array)):
    if remainder_by_3[i] == remainder_by_7[i]:
      list_vide.append(i)

  # création d'une liste vide pour stocker les valeurs dont les %3 et %7 sont égales
  list_of_int = []
  for i in range(len(list_vide)):
    list_of_int.append(array[list_vide[i]])
  return list_of_int

array = np.arange(-100, 101)

remainder_euclidean(array)
```

## Mission 3

Create a function that returns the [inverse value (for `3`, return `1/3`)](https://en.wikipedia.org/wiki/Multiplicative_inverse) of a set of values contained in a NumPy array

```
def reciproque(array):
  inversed_array = np.power(array, -1)
  return inversed_array


my_array = np.arange(2, 20, dtype=float)
my_array
```

```
reciproque(my_array)
```
