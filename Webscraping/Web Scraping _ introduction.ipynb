4. Trouve la balise permettant d'identifier le bloc contenant une blague. Combien y en a-t-il ? (il devrait y en avoir 20). Affiche le 8e bloc de blague grâce à l'attribut .text

```python
# object beautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')

# find the right 'balise'
contenu = soup.find(id='content')

# the jokes don't have 'class or id balise' : let's find the <p> in the <div id = 'content'>
jokes = contenu.find_all('p')

# how many jokes ?
len(jokes)
```

On pouvait aussi partir de la balise avec le style CSS mais le .text ne rendait pas aussi bien :

soup.find_all('div', {"style" : "border-top: 1px solid #251836; padding: 0 0 5px 7px;"})

```python
# 8th joke :
jokes[7].get_text()
```

6. Fais de même pour isoler uniquement la note de la 8e blague

```python
notes = contenu.find_all(class_ ='out5Class')
notes[7].get_text()
```

7. Crée un dictionnaire vide, puis crée une boucle qui viendra enregistrer dans ce dictionnaire chaque blague en clé et chaque note correspondante en valeur.

```python
jokes_list=[]
for joke in jokes :
    jokes_list.append(joke.get_text())

notes_list= []
for note in notes:
    notes_list.append(note.get_text())

jokes_notes_dict = dict(zip(jokes_list, notes_list))
jokes_notes_dict
```

8. Transforme ce dictionnaire en DataFrame de 2 colonnes : blague et note. Il doit avoir 20 lignes : une par blague.

```python
df = pd.DataFrame({
    'Blagues' : jokes_notes_dict.keys(),
    'Note' : jokes_notes_dict.values()
})

df
```
