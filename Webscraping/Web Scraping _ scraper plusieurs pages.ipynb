```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import plotly.express as px
```

1. Observe attentivement l'URL de la page 1. Essaye de passer à la page 2. Dans un notebook python, essaye de générer automatiquement les URL des 5 premières pages (avec une boucle par exemple ?)

```python
pages = []

for i in range (1,6):
  url = 'https://chucknorrisfacts.net/facts.php?page=' + str(i)
  pages.append(url)
```

2. Utilise BeautifulSoup pour scraper les facts et les notes des 5 premières pages, grâce aux URL que tu as générées. (Nous pourrions faire les 100 ou 200 pages suivantes, mais ca risque d'entrainer une charge inutile pour ceux qui administrent ce serveur.) Le résultat doit être dans un dictionnaire.

```python
# for User-Agent:
nav = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.93 Safari/537.36"

jokes_dict ={}
for url in pages:
  # requests
  r = requests.get(url, headers={'User-Agent': nav})
  #soup
  soup = BeautifulSoup(r.text, 'html.parser')
  
  #dict creation
  jokes =soup.find_all('div',{'style' : 'border-top: 1px solid #251836; padding: 0 0 5px 7px;'})
  for i in range(20):
    key = jokes[i].find('p').get_text()
    note = jokes[i].find('span').get_text()
    jokes_dict[key] = note
```

```python
# il doit y avoir 100 entrées logiquement
len(jokes_dict)
```

3. Crée un DataFrame contenant ces données, il doit donc posséder 2 colonnes et 100 lignes.

```python
df = pd.DataFrame({
    'Blagues' : jokes_dict.keys(),
    'Note' : jokes_dict.values()
})

df
```

4. Crée une visualisation de la distribution des notes.

```python
fig = px.histogram(df, x='Note', title="Distribution of Chuck Norris jokes' notes", color_discrete_sequence=['indianred'] )
fig.update_layout(title ={'x': 0.5}, autosize=False)
fig.show()
```
