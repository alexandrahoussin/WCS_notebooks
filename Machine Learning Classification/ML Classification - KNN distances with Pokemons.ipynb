```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.neighbors import NearestNeighbors
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/murpi/wilddata/master/pokemon.csv', index_col=0)
df.head()
```

```python
# DF avec les pokemons non légendaires
pokedex = df[df['Legendary'] == False] 
```

```python
# df avec seulement mes pokemons
my_poke = ['Mewtwo', 'Lugia', 'Rayquaza', 'Giratina Origin Forme', 'Dialga', 'Palkia']
my_poke_df = df[df['Name'].isin(my_poke)]
my_poke_df
```

```python
# ne garder que les colonnes numériques
colonnes =  list(pokedex.columns)
colonnes = colonnes[3:-1]

# définir X :
X = pokedex[colonnes]

# entrainement
distknn = NearestNeighbors(n_neighbors= 1).fit(X)
```

```python
preco_poke = distknn.kneighbors(my_poke_df[colonnes])
```

```python
preco_poke = preco_poke[1].reshape(1,6)[0]

preco_poke
```

```python
# df des preco pokemon
df_preco = pokedex.iloc[preco_poke]
df_preco
```

```python
for i in range(len(my_poke)):
  mine = my_poke[i]
  remplacement = df_preco['Name'].iloc[i]
  print(f"Reviens {mine}, à toi {remplacement} ! ")
```
