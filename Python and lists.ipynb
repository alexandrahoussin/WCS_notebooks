# Mission 1

Create a script that takes a list and displays only the elements (and their position) that have positions that are even (0, 2, 4, 6...).

Example list = ["a", "b", "c", "d", "e"], the script should return : 
- a position 0
- c position 2
- e position 4

```
my_list = ["a", "b", "c", "d", "e"]

for i, element in enumerate(my_list):
  if i % 2 == 0:
    print('{} position {}'.format(element, i))
```

# Mission 4

Create a script that does the same thing as the insert method, but at a random position.

```
# import de randrange pour générer un index aléatoire
from random import randrange

# liste initiale
my_list = ['R2D2', 'Luc', 'C3PO', 'Leia']

# inserer 'Han' dans my_list a un index aléatoire
my_list.insert(randrange(len(my_list)+1), 'Han')

print(my_list)
# si vous lancez le code plusieurs fois, Han sera a des positions différentes
```

# Mission 5

Create a script that does the same thing as the pop method for a list containing only characters. The difference is that this function displays a message indicating the number of characters that are identical to the deleted character and still remain in the list. In other words, if there were 3 "b" in the list, and one has just been deleted by your function, the script will display "there are still 2 letters b".

```
my_list = ['a', 'a', 'c', 'b', 'b', 'c', 'd', 'd', 'a','b', 'c']

# choix de l'index de la lettre à supprimer
index = int(input("Choisissez l'index du caractère à supprimer (maximum : {}): ".format(len(my_list))))

# assigne cette lettre à ma variable letter
letter = my_list.pop(index)

# utilisation de la méthode count() pour compter le nombre de lettre restantes
print('There are still {} letters {}.'.format(my_list.count(letter), letter))
```

```
# en choisissant la lettre
my_list = ['a', 'a', 'c', 'b', 'b', 'c', 'd', 'd', 'a','b', 'c']

letter_to_remove = input('Choisissez la lettre à supprimer: ')

if letter_to_remove not in my_list:
  print("Cette lettre n'est pas dans la liste")
else:
  index_to_remove = my_list.index(letter_to_remove)
  my_list.pop(index_to_remove)

  print('il y a toujours {} lettres {}.'.format(my_list.count(letter_to_remove), letter_to_remove))
```

# Mission 6

Create a script with 3 lists, and indicates the two that are "closest". That is to say, the two that have the most elements in common.

```
import numpy as np

list1 = ['hello', 'bonjour', 'hola', 'bye']
list2 = ['bonjour', 'ciao', 'bye']
list3 = ['salut', 'ciao', 'bonjour', 'bye']

# utilisation np.intersect1d qui renvoie les caractères identiques de deux listes, puis utilisation de len pour calculer le nombre
one_two = len(np.intersect1d(list1, list2))
one_three = len(np.intersect1d(list1, list3))
two_three = len(np.intersect1d(list2, list3))


if one_two > one_three and one_two > two_three:
  print('list1 and list2 are closest')
elif one_three > one_two and one_three >two_three:
  print('list1 and list3 are closest')
elif one_two == one_three and one_two == two_three:
  print("the three lists have the same number of elements in common")
else:
  print('list2 and list3 are closest')
```
